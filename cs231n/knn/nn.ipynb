```python
def l1_distance(img1, img2):
  diff = img1 - img2
  return np.sum(np.abs(diff))
```

```python
print(l1_distance(img1, img2))
```

```python
def l2_distance(img1, img2):
  diff = img1 - img2
  return np.sqrt(np.sum(np.square(diff)))
```

```python
print(l2_distance(img1, img2))
```

```python
X_train = np.array([
    [1, 1],
    [2, 2],
    [3, 3]
])

y_train = np.array([
    "cat",
    "dog",
    "mug"
])
```

```python
nearest = np.argmin([l1_distance(img, X_test) for img in X_train])

print(y_train[nearest])
```

```python
import numpy as np


class NearestNeighbor(object):

    def __init__(self):
        """
        Constructor.

        This runs when:
            nn = NearestNeighbor()

        At this point the model knows nothing.
        """
        pass


    def train(self, X, y):
        """
        X : Training images
            Shape = (N, D)

        y : Labels
            Shape = (N,)

        N = number of training examples
        D = number of features/pixels

        Example:
            X.shape = (50000, 3072)
            y.shape = (50000,)
        """

        # Simply memorize training data
        self.Xtr = X
        self.ytr = y


    def predict(self, X):
        """
        Predict labels for test data.

        X:
            Shape = (num_test, D)

        Returns:
            Ypred
        """

        num_test = X.shape[0]

        Ypred = np.zeros(
            num_test,
            dtype=self.ytr.dtype
        )

        # Loop through every test image
        for i in range(num_test):

            # Compute L1 distance between
            # current test image and ALL training images
            distances = np.sum(
                np.abs(
                    self.Xtr - X[i, :]
                ),
                axis=1
            )

            # Find smallest distance
            min_index = np.argmin(
                distances
            )

            # Copy label of nearest image
            Ypred[i] = self.ytr[
                min_index
            ]

        return Ypred
```
